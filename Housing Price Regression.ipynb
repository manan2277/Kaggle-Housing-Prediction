```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'mean')),
    ('std_scaler', StandardScaler())   
])
cat_pipeline = Pipeline([
    ('cat_imputer', SimpleImputer(strategy = 'most_frequent')),
    ('OH_encoder', OneHotEncoder(handle_unknown='ignore'))
])
#Using a simple imputer here with strategy = mean, could have used something more complex especially for lot_frontage
#with ~200 entries missing
full_pipeline = ColumnTransformer([('num', num_pipeline, num_cols), 
                                  ('cat', cat_pipeline, cat_cols)])
X = df.drop(['Id', 'SalePrice'], axis=1)
Y = df.SalePrice.copy()
X_test = df_test.drop(['Id'] ,axis=1)
```

```python
# Processing our data through the pipeline
X = full_pipeline.fit_transform(X)
X_test = full_pipeline.transform(X_test)
```
